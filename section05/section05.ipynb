## ここでやること

- 発言データをどの単位で扱うか理解する

- API取得やQA生成につながる流れを把握する


```
import requests

url = "https://kokkai.ndl.go.jp/api/meeting_list"

params = {
    "recordPacking": "json",
    "maximumRecords": 1,
    "nameOfMeeting": "予算委員会"
}

res = requests.get(url, params=params).json()

meeting = res["meetingRecord"][0]

print("会議名:", meeting["nameOfMeeting"])
print("issueID:", meeting["issueID"])
```

```
issue_id = meeting["issueID"]

url = "https://kokkai.ndl.go.jp/api/speech"

params = {
    "recordPacking": "json",
    "issueID": issue_id,
    "maximumRecords": 5
}

res = requests.get(url, params=params).json()

for s in res["speechRecord"]:
    print("発言者:", s["speaker"])
    print("内容:", s["speech"][:50], "...")
    print("------")
```
